we extract total no of floors 

```python
def extract_total(x):
    x = str(x)
    if "out of" in x:
        return int(x.split('out of')[1].strip())
    else:
        return None
df['Total Floors'] = df['Floor'].apply(extract_total)
```

```python
df['Total Floors'].isnull().sum()
```

```python
df['Total Floors'].fillna(df['Total Floors'].median(), inplace=True)
```

```python
df['Total Floors'].astype(int)
```

We do OHE  on remainng column to convert it into  number

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder4 = OneHotEncoder()
encoder5= OneHotEncoder()
encoded1  = encoder.fit_transform(df[['Furnishing Status']]).toarray()
encoded2  = encoder4.fit_transform(df[['City']]).toarray()      
encoded3  = encoder5.fit_transform(df[['Area Type']]).toarray()
```

```python
df.head()
```

```python
enc1_df = pd.DataFrame(encoded1,columns=encoder.get_feature_names_out())
enc2_df = pd.DataFrame(encoded2,columns=encoder4.get_feature_names_out())
enc3_df = pd.DataFrame(encoded3,columns=encoder5.get_feature_names_out())
df = pd.concat([df,enc1_df,enc2_df,enc3_df],axis=1)
df.drop(['Furnishing Status', 'City', 'Area Type'], axis=1, inplace=True)
```

```python
df.drop('Floor',axis=1 ,inplace=True)
df.head()
```
